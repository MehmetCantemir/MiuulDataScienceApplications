# GÖREV 3

Çıktıyı PRICE'a göre sıralayınız.

- Önceki sorudaki çıktıyı daha iyi görebilmek için sort_values metodunu azalan olacak şekilde PRICE'a uygulayınız.
- Çıktıyı agg_df olarak kaydedin.

```python
agg_df=df.groupby(["COUNTRY","SOURCE","SEX","AGE"]).agg({"PRICE":"mean"}).sort_values("PRICE",ascending=False)
agg_df.head(10)
```

# GÖREV 4

**Indekste yer alan isimleri değişken isimlerine çeviriniz.**

- Üçüncü sorunun çıktısında yer alan PRICE dışındaki tüm değişkenler index isimleridir. Bu isimleri değişken isimlerine çeviriniz.



```python
agg_df=agg_df.reset_index()
agg_df.head(10)
```

# GÖREV 5

**Age değişkeninin kategorik değişkene çeviriniz ve agg_df'e ekleyin**

- Age sayısal değişkenini kategorik değişkene çevirin
- Aralıkları ikna edici şekilde oluşturunuz
- Örneğin : "0_10" , "19_23" , "24_30" ... gibi

```python
# AGE değişkenin nerelerden bölüneceğini belirleme :
bins =[0,10,23,30,40,agg_df["AGE"].max()]

# Bölümün noktalara karşılık isimlendirmelerin ne olacağını ifade edelim
mylabels=['0_18','19_23','24_30','31_40','41_'+str(agg_df["AGE"].max())]

# age'i bölelim
agg_df["age_cat"]=pd.cut(agg_df["AGE"],bins,labels=mylabels)
agg_df.head(10)
```

# GÖREV 6

**Yeni seviye tabanlı müşterileri tanımlayınız.**
- Yeni seviye tabanlı müşterileri tanımlayın ve veri setine değişken olarak ekleyin
- Yeni eklenecek değişkenin adı : custormes_lebel_based
- Önceki soruda elde edeceğini çıktıdaki gözlemleri bir araya getirerek customers_level_based değişkenini oluşturmanız gerekmektedir.

```python
agg_df["customers_label_based"]=[row[0].upper() + "_" + row[1].upper() + "_" + row[2].upper() + "_" + row[5].upper() for row in agg_df.values]
agg_df.head(10)
```

# GÖREV 7

**Yeni müşterileri segmentlere ayırınız.**

- Yeni müşterileri PRICE'a göre 4 segmente ayırınız.
- Segmentleri SEGMENT isimlendirmesi ile değişken olarak agg_df'e ekleyiniz.
- Segmentleri betimleyiniz.(mean , max,sum'larını alınız)

```python
agg_df["SEGMENT"]=pd.qcut(agg_df["PRICE"],4,labels=["D","C","B","A"])
agg_df.head(10)
```

```python
agg_df.groupby("SEGMENT").agg({"PRICE":["mean","sum","max"]})
```
