# GÖREV 2

SaleCheckIndDayDiff değişkeni müşterinin CheckIn tarihinden ne kadar önce satın alımınını tanımladığını gösterir.

Aralıkları ikna edici şekilde oluşturunuz.

Örneğin : '0_7' , '7_30', '30-90', '90_max' aralıkları kullanılabilir.

Bu aralıklar için "Last Minutes" , "Potential Planners" , "Planners" , "Early Bookers"

```python
bins =[-1,7,30,90,df["SaleCheckInDayDiff"].max()]
labels = ["Last Minutes", "Potential Planners", "Planners" , "Early Bookers"]

df["EB_Score"] = pd.cut(df["SaleCheckInDayDiff"], bins, labels=labels)
df.head(50).to_excel("eb_score.xlsx", index=False)
```

# GÖREV 4

**City - Concept - Season kırılımının çıktısını PRICE'e göre sıralayınız**

```python
agg_df=df.groupby(["SaleCityName","ConceptName","Seasons"]).agg({"Price":"mean"}).sort_values("Price",ascending=False)
agg_df.head(20)
```

# GÖREV 5

**Dördüncü sorunun çıktısında yer alan PRICE dışındaki tüm değişkenler index isimleridir. Bu isimleri değişken isimlerine çeviriniz.**



```python
agg_df.reset_index(inplace=True)
agg_df.head(10)
```

# GÖREV 6

- Yeni seviye tabanlı satışları tanımalyınız ve veri setine değişken olarak ekleyiniz.

- Yeni eklenecek değişkenin adı: sales_level_based

- Önceki soruda elde edeceğiniz çıktıdaki gözlemleri bir araya getirererk sales_level_Based değişkenini oluşturmanız gerekir.

```python

agg_df['sales_level_based']=agg_df[["SaleCityName","ConceptName","Seasons"]].agg(lambda x : '_'.join(x).upper(),axis=1 )

agg_df.head(10)
```

# GÖREV 7

- Yeni personaları PRICE'e göre 4 segmente ayırınız

- Segmentleri SEGMENT isimlendirmesi ile değişken olarak agg_df'e ekleyiniz

- Segmentleri betimleyiniz(Segmentlere göre group by yapıp price mean , max ,sum'larını alınız)

```python
agg_df["SEGMENT"]=pd.qcut(agg_df["Price"],4,labels=["D","C","B","A"])
agg_df.head(100)
```

```python
agg_df.groupby(["SEGMENT"]).agg({"Price":["mean","sum","max"]})
```
